# Bar Plot

```python
df.day.unique()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

sns.barplot(x='day',y='tip',data=df);
```

```python
sns.barplot(x='day',y='total_bill',data=df);
```

```python
plt.figure(figsize=(8,5))

#plt.title("Bar Plot")
sns.barplot(x='day',y='tip',data=df,hue='sex',palette='winter_r');
```

```python
plt.figure(figsize=(8,5))

#plt.title("Bar Plot")
sns.barplot(x='day',y='total_bill',data=df,hue='smoker',palette='winter_r');
```

```python
sns.barplot(x='day',y='total_bill',data=df,hue='smoker',palette='winter_r');
plt.xlabel("Name of Days");
plt.title("Average Total Bill based on Smoking habits");
```

# joint plots

```python
#jointplot helps to do bivariate analysis. It provides visualization of relationship between two variables
ax=sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='kde');
```
