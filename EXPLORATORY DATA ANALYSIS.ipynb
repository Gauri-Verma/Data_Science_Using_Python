# Using percentile med to detec outliers

```python
# example knowing about a dataset which lies above 95% of quantile

max_thresold = df.height.quantile(0.95)
max_thresold

# output tells us that anything above this value i.e. 9.68 will be considered as outlier
```

```python
# finding datasets which are heigher then threshold value

df[df['height']>max_thresold]                    #these will be outliers
```

```python
min_thresold = df['height'].quantile(0.05)
min_thresold

# output tells us that any value below 3.65 will be considered as an outlier
```

```python
# finding datasets which are less then threshold value

df[df['height']<min_thresold]               #outlier
```

```python
#Remove outliers

df1=df[(df['height']<max_thresold) & (df['height']>min_thresold)]
df1                 #dataset without outliers
```

```python
import seaborn as sns

fig = plt.figure(figsize =(7, 7)) 
sns.boxplot(df1['height'], orient='v')         #boxplot after removing outliers 
```

# Data Normalization

Data Normalization
Normalization is a rescaling of the data from the original range so that all values are within the new range of 0 and 1.

A value is normalized as follows:

y = (x – min) / (max – min)

```python
# example of a normalization
#from numpy import 
import numpy as np
from sklearn.preprocessing import MinMaxScaler
# define data
data = np.array([[100, 0.001],
				[8, 0.05],
				[50, 0.005],
				[88, 0.07],
				[4, 0.1]])
print(data)
# define min max scaler
scaler = MinMaxScaler()
# transform data
scaled = scaler.fit_transform(data)
print('\n',scaled)
```

Data Standardization

This can be thought of as subtracting the mean value or centering the data.

A value is standardized as follows:

y = (x – mean) / standard_deviation
Where the mean is calculated as:

mean = sum(x) / count(x)
And the standard_deviation is calculated as:

standard_deviation = sqrt( sum( (x – mean)^2 ) / count(x))

```python
# example of a standardization

from sklearn.preprocessing import StandardScaler
# define data
data = np.array([[100, 0.001],
				[8, 0.05],
				[50, 0.005],
				[88, 0.07],
				[4, 0.1]])
print(data)
# define standard scaler
scaler = StandardScaler()
# transform data
scaled = scaler.fit_transform(data)
print('\n',scaled)
```
