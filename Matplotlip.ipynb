```python
# how to give label/legends
a1 = [5,8,10]
b1 = [12,16,6]

a2 = [6,9,11]
b2 = [6,15,7]

plt.plot(a1, b1, 'r', label = 'line 1', linewidth = 8)
plt.plot(a2, b2, 'b', label = 'line 2', linewidth = 4)

plt.title('Epic Info')  #graph title
plt.ylabel('Y axis')    #Y axis title
plt.xlabel('X axis')    #X axis title

plt.legend(loc='lower right')
#plt.legend()
plt.show()
```

# barchart

```python
x=[2,8,10]
y=[11,15,9]

x2=[12,5,15]
y2=[6,8,10]

plt.bar(x,y, color='c')
plt.bar(x2,y2,color='g')
plt.title("Bar plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
plt.figure(figsize=(8,8))
plt.bar([1, 3, 5, 7, 9], [5, 2, 7, 8, 2], label = 'Instance 1', color='g')
plt.bar([2, 4, 6, 8, 10], [8, 6, 2, 5, 6], label = 'Instance 2', color ='r')

plt.legend()

plt.ylabel('Bar number')
plt.xlabel('Bar height')

plt.title("Bar plot")

plt.show()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
plt.bar(company,revenue,label='revenue', color='b')
plt.legend()
plt.xlabel("Stocks")
plt.ylabel("revenue")
```

```python
#plotting revenue and profit side by side
import numpy as np

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

xpos=np.arange(len(company))

plt.xticks(xpos,company)
#plt.xticks(xpos,company,rotation='90')
plt.bar(xpos-0.2,revenue,width=0.3,label='revenue',color='red')
plt.bar(xpos+0.2,profit,color='blue',width=0.3,label='profit')
plt.legend()
plt.xlabel("Stocks")
plt.ylabel("revenue and profit")
```

# Histogram

```python
data=np.random.randint(0,10,20)
print(data)
plt.hist(data);
```

```python
plt.figure(figsize=(8,8))
data=np.random.randint(0,70,100)
plt.hist(data,bins=20,histtype='stepfilled',color='y',alpha=0.8); #alpha will increase or decrease shade of colour
```

# Scatter Plot

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 2, 1, 4, 5, 2]

plt.scatter(x, y, label = 'cat', color = 'k', s=160 , marker = "*")  # s is marker size
#plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")

plt.legend()
plt.show()
```

# Pie Charts

```python
fractions = [7, 2, 2, 13]
activities = ['sleeping', 'eating', 'coding', 'playing']
cols = ['g', 'm', 'r', 'b']

#plt.pie(fractions)

plt.pie(fractions , labels = activities, colors = cols, startangle = 90, shadow = True, explode = (0, 0.1, 0.2, 0), autopct = '%1.1f%%')

plt.title('Pie Plot')
plt.show()
```
